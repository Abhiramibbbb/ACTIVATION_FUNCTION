```python
import numpy as np
import matplotlib.pyplot as plt

# Define different activation functions
def step_function(x):
    return 1 if x >= 0 else 0

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return max(0, x)

# XOR function with customizable activation function
def XOR(ip, input_weights1, input_weights2, hidden_bias, hidden_weights, output_bias, activation_function):
    # Compute hidden layer activations
    n1 = ip[0]*input_weights1[0] + ip[1]*input_weights2[0] + hidden_bias[0]
    h1 = activation_function(n1)
    
    n2 = ip[0]*input_weights1[1] + ip[1]*input_weights2[1] + hidden_bias[1]
    h2 = activation_function(n2)
    
    # Compute output layer activation
    output = h1*hidden_weights[0] + h2*hidden_weights[1] + output_bias
    out = activation_function(output)
    
    # For step function, convert output to binary (0 or 1)
    if activation_function == step_function:
        out = step_function(output)
    
    return out

# Get input from user
ip = list(map(float, input("Enter the inputs (e.g., 1 1): ").split()))
input_weights1 = list(map(float, input("Enter the weights from input 1 to hidden layer (e.g., 1 -1): ").split()))
input_weights2 = list(map(float, input("Enter the weights from input 2 to hidden layer (e.g., 1 -1): ").split()))
hidden_bias = list(map(float, input("Enter the biases for the hidden layer (e.g., 0 0): ").split()))
hidden_weights = list(map(float, input("Enter the weights from hidden layer to output (e.g., 1 1): ").split()))
output_bias = float(input("Enter the bias for the output node (e.g., -1.5): "))

# Choose activation function
activation_choice = input("Choose activation function (step/sigmoid/tanh/relu): ").strip().lower()
if activation_choice == "step":
    activation_function = step_function
elif activation_choice == "sigmoid":
    activation_function = sigmoid
elif activation_choice == "tanh":
    activation_function = tanh
elif activation_choice == "relu":
    activation_function = relu
else:
    print("Invalid activation function chosen. Defaulting to step function.")
    activation_function = step_function

# Compute the XOR output
perceptron = XOR(ip, input_weights1, input_weights2, hidden_bias, hidden_weights, output_bias, activation_function)
print(f"The output of the XOR function is: {perceptron}")

# Plotting activation functions
def plot_activation_function(x, y, title):
    plt.plot(x, y)
    plt.title(title)
    plt.xlabel("Input")
    plt.ylabel("Output")
    plt.grid(True)
    plt.show()

x_values = np.linspace(-10, 10, 400)
y_step = [step_function(x) for x in x_values]
y_sigmoid = sigmoid(x_values)
y_tanh = tanh(x_values)
y_relu = [relu(x) for x in x_values]

# Plotting Step Function
plot_activation_function(x_values, y_step, "Step Function")

# Plotting Sigmoid Function
plot_activation_function(x_values, y_sigmoid, "Sigmoid Function")

# Plotting Tanh Function
plot_activation_function(x_values, y_tanh, "Tanh Function")

# Plotting ReLU Function
plot_activation_function(x_values, y_relu, "ReLU Function")
```
